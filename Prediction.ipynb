```python
#!/usr/bin/env python
# coding: utf-8

# In[22]:


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import urllib.request
import json
import time


dataset=pd.read_csv('HeatDemand3year.csv')
dataset.dropna
dataset = dataset[(dataset['Outdoor Temperature'] <= 20) & (dataset['DH'] <= 300)]
outdoor_temp = dataset['Outdoor Temperature'].values.reshape(-1,1)
dh = dataset['DH'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(outdoor_temp,dh, test_size=0.2, random_state=0)
Model = LinearRegression()
Model.fit(X_train, y_train)
Model.score(X_train, y_train)

while (1):
	try:
		y_pred4 = Model.predict(X_test)
		r2_score(y_test, y_pred4)


		plt.scatter(X_test, y_test,  color='blue')
		plt.plot(X_test, y_pred4, color='#3bfb00', linewidth=2)
		plt.show()


		f = urllib.request.urlopen('http://dataservice.accuweather.com/forecasts/v1/hourly/12hour/330088?apikey=BkXoKvLAOel6syGdDYCOPKyJiXqOFA5a')
		response = json.loads(f.read())
		weatherforecast = np.array([])
		for i in range(12):
		    temp=(response[i]['Temperature']['Value']-32)*5/9
		    weatherforecast = np.append(weatherforecast,temp) 

		Forecast_DH = Model.predict(weatherforecast.reshape(-1,1))

		plt.scatter(weatherforecast, Forecast_DH,  color='blue')
		plt.show()


		Forecast_DH = pd.DataFrame(Forecast_DH.round(2),columns=['prediction'])
		for i in Forecast_DH.prediction:
		    url = 'http://dweet.io/dweet/for/DH_Prediction?temp=' + str(i)
		    send_p = urllib.request.urlopen(url)
		    response1 = json.loads(send_p.read())
		    print(url)

		time.sleep(3600)
	except:
		print ("Error")
		time.sleep(100)
    






```
